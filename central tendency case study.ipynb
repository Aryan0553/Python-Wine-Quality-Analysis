```python
#histogram of fixed acidity feature
plt.figure(figsize=(11,6))

sns.histplot(data=df, x= 'fixed acidity', color = 'orange', 
             edgecolor='linen', alpha = 0.5, bins = 5)
plt.title("Histogram of fixed acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')
plt.show()
```

```python
#calculated the mean of fixed acidity
round(df['fixed acidity'].mean(),2)
```

```python
#median
df['fixed acidity'].median()
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data=df, x='fixed acidity', color='orange', 
             edgecolor='linen', alpha=0.5, bins=5)
plt.title("Histogram of fixed acidity")
plt.xlabel('Fixed Acidity')
plt.ylabel('Count')

plt.vlines(df['fixed acidity'].mean(), ymin=0, ymax=4000, colors='blue', label='Mean')
plt.vlines(df['fixed acidity'].median(), ymin=0, ymax=4000, colors='red', label='Median')

plt.legend()
plt.show()
```

```python
#craete a volatile acidity feature 
plt.figure(figsize=(11,6))

sns.histplot(data=df, x='volatile acidity', color='green', 
             edgecolor='linen', alpha=0.5, bins=5)
plt.title("Histogram of volatile acidity")
plt.xlabel('volatile acidity')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(11,6))

sns.distplot(df['volatile acidity'],color='purple')
plt.title(" Distplot of volatile acidity")
plt.xlabel('volatile acidity')
plt.ylabel('Density')
plt.show()
```

```python
#calculate skweness function of python 
df['volatile acidity'].skew()
```

```python
plt.figure(figsize=(11,6))

sns.histplot(data=df, x='volatile acidity', color='green', 
             edgecolor='linen', alpha=0.5, bins=5)
plt.title("Histogram of volatile acidity")
plt.xlabel('volatile acidity')
plt.ylabel('density')

plt.vlines(df['volatile acidity'].mean(), ymin=0, ymax=4000, colors='blue', label='Mean')
plt.vlines(df['volatile acidity'].median(), ymin=0, ymax=4000, colors='red', label='Median')

plt.legend()
plt.show()
```

```python
#citric acid histogram 
plt.figure(figsize=(11,6))

sns.histplot(data=df, x= 'citric acid', color = 'red', 
             edgecolor='linen', alpha = 0.5, bins = 5)
plt.title("Histogram of citric acidy")
plt.xlabel('citric acidy')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(11, 6))

# Generate the count plot with corrected syntax
sns.countplot(x='quality', data=df)

# Add labels and title with quotes
plt.title('Count plot of quality')
plt.xlabel('quality')
plt.ylabel('Count')

# Display the plot
plt.show()
```
